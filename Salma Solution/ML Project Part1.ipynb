## feature ecoding to do correlation


```python
from sklearn import preprocessing
# encode title column
le = preprocessing.LabelEncoder()
data_frame['title'] = le.fit_transform(data_frame['title'])
```

```python
# encode authors column
data_frame['authors'] = le.fit_transform(data_frame['authors'])
```

```python
# encode language column
enc_lang = pd.get_dummies(data_frame['language_code'])
data_frame = pd.concat([data_frame, enc_lang], axis = 1)
```

```python
# Encode Publication Date
data_frame['publication_year'] = pd.to_datetime(data_frame['publication_date'], format='%m/%d/%Y',errors='coerce').dt.year
data_frame['publication_month'] = pd.to_datetime(data_frame['publication_date'], format='%m/%d/%Y',errors='coerce').dt.month
data_frame['publication_day'] = pd.to_datetime(data_frame['publication_date'], format='%m/%d/%Y',errors='coerce').dt.day

# Drop the original publication_date column
data_frame.drop('publication_date', axis=1, inplace=True)

# Encode Publisher (Assuming you're using LabelEncoder)
le = preprocessing.LabelEncoder()
data_frame['publisher'] = le.fit_transform(data_frame['publisher'])
data_frame['language_code'] = le.fit_transform(data_frame['language_code'])
```
