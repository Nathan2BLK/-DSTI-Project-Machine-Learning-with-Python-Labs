import json
import csv

# Specify the JSON file name
json_file_path = '/Users/yosrnoureddine/Downloads/goodreads_book_authors.json'
csv_file_name = 'authors.csv'

# Open the CSV file for writing
with open(csv_file_name, 'w', newline='') as csv_file:

    # Create a CSV writer
    csv_writer = csv.writer(csv_file)

    # Open the JSON file for reading
    with open(json_file_path, 'r') as json_file:

        # Read and process each line in the JSON file
        for line in json_file:

            try:
                # Load each line as a JSON object
                json_data = json.loads(line)

                # If the CSV file is empty, write the header
                if csv_file.tell() == 0:
                    header = json_data.keys()
                    csv_writer.writerow(header)

                # Write the values to the CSV file
                csv_writer.writerow(json_data.values())

            except json.JSONDecodeError as e:
                print(f"Error decoding JSON: {e}")

# Provide a message indicating the CSV file has been generated
print(f"CSV file '{csv_file_name}' has been generated.")
