```python
# creating a separate dataset with only numerical values to create a correlation matrix
# keeping only the 5 main languages of the dataset : english, french, german, japanese, spanish
df_corr = df_processed.drop(['title', 'authors', 'isbn', 'language_code', 'publication_date','publisher','mul','zho','grc','por','ita','lat','enm','rus','swe','nl','ara','srp','msa','glg','wel','nor','tur','gla'], axis=1)
```

```python
df_corr.head(10)
```

```python
corr_matrix = df_corr.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

We can notice:
* Most of the features are poorly correlated, except for the ratings_count and text_reviews_count which are strongly correlated.
* The average_rating is poorly correlated to all features. The feature that correlates the most is the num_pages.
